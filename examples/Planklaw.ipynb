```julia
using RadioTools
using PyPlot
```

```julia
start = 8     #order of magnitude (100 Mhz)
stop  = 15    #order of magnitude (1e3 Thz)
n     = 1000  #elements number
ν     = logspace(start, stop, n)
t_eff = 5772  #Temperature
Bν    = plank(ν, t_eff);
```

```julia
plot(ν, Bν)
title("Plank law (T = $t_eff K)")
xlabel("Frecuency (Hz)")
ylabel("Bν(T) (erg s⁻¹ Hz⁻¹ cm⁻² sr⁻¹)")

show()
```

```julia
loglog(ν,Bν)
title("Plank law (T = $t_eff K)")
xlabel("Frecuency (Hz)")
ylabel("Bν(T) (erg s⁻¹ Hz⁻¹ cm⁻² sr⁻¹)")

show()
```

```julia
n2  = 50
ν2  = 12e9
T2  = t_eff
I02 = 1.0

Iν  = intensity(n2, ν2, T2, I02)
plot(Iν)
title("Intensity at 12GHz")
xlabel("Step")
ylabel("Intensity")

show()
```
